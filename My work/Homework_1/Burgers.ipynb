## Ex. 3 - Burgers' eq
### Conservative form:

```python
import numpy as np
import csv

run_name = 'J_101' #J_501'

# Define the domain
L = 10.0     # Domain length
nx = 101    # Number of grid points
dx = L/(nx-1)   # Grid spacing
x = np.linspace(0, L, nx)


# Define the initial condition
x0=5
u_initial = 10*np.exp(-(x-x0)**2)

# Initialize the solution array
u_current = u_initial.copy() 

# Define the time step and the final time
cf = 0.5     #Courant factor
dt = cf*dx/(np.max(u_current))   #Time step
t_final = 0.5

print('nx=',nx)
print('dx=',dx)
print('dt=',dt)
print('Number of iterations=',t_final/dt)

#we create arrays where we will store the time and the l2norm
l2norm=[]
l2norm.append(np.sqrt(np.sum(u_current**2)/len(u_current))) #We expect it to be a consant! 

time=[]
time.append(0.0)
```

```python
import matplotlib.pyplot as plt
import os
##create directory where to save images
print(os.getcwd())

os.makedirs('./images_' + str(run_name))

os.listdir('./')
```

```python
# Initilize time and iteration counter
t = 0.0
i = 0

#save the initial conditions
plt.plot(x, u_current)
plt.title('Time='+str(round(t,2)))
plt.grid(c = 'gray', alpha = 0.3)
plt.xlabel('x')
plt.ylabel('u')
plt.ylim(0,10.5)
plt.savefig('./images_' + str(run_name) + '/fig_'+str(i).zfill(5)+'.png', dpi=200)
plt.close()

#solve the advection equation
while t < t_final:
    u_next = u_current - (dt/dx)*(0.5*(u_current)**2 - 0.5*((np.roll(u_current, 1))**2))    
    
    # Update the solution
    u_current = u_next.copy()
    
    #Update the time step with the new velocity
    dt = cf*dx/(np.max(u_current))
    
    #advance the time 
    t += dt
    i += 1
    
    #compute the l2 norm and add the time to the time vector
    l2norm.append(np.sqrt(np.sum(u_current**2)/len(u_current)))
    time.append(t)
    
    #plot the current result and save in an image every 10 iterations
    if (i%5==0):
        plt.plot(x, u_current)
        plt.grid(c = 'gray', alpha = 0.3)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title('Time='+str(round(t,2)))
        plt.ylim(0,10.5)
        plt.savefig('./images_' + str(run_name) + '/fig_'+str(i).zfill(5)+'.png', dpi=200)
        plt.close()
```

```python
# Plot the final solution
plt.plot(x, u_initial, label='Initial')
plt.plot(x, u_current, label='Final')
plt.title(str(run_name))
plt.ylim(0,10.5)
plt.xlabel('x')
plt.ylabel('u')
plt.grid(color = 'gray', alpha = 0.3)
plt.legend()
#plt.show()
plt.savefig('i_f_burg_'+str(run_name)+'.png', dpi = 300)
```

```python
#Plot the L2 norm
plt.plot(time,l2norm/l2norm[0])
plt.ylabel(r'$L_2$ norm')
plt.xlabel('Time')
plt.grid(c = 'gray', alpha = 0.3)
plt.show()
```

### Non-conservative Form:

```python
#Set the name of the run:
run_name_NCF = 'NCF_101' #'NCF_501'

# Define the domain
L = 10.0     # Domain length
nx = 101    # Number of grid points
dx = L/(nx-1)   # Grid spacing
x = np.linspace(0, L, nx)


# Define the initial condition
x0=5
u_initial = 10*np.exp(-(x-x0)**2)

# Initialize the solution array
u_current_NCF = u_initial.copy() 

# Define the time step and the final time
cf = 0.5     #Courant factor
dt = cf*dx/(np.max(u_current_NCF))   #Time step
t_final = 0.5

print('nx=',nx)
print('dx=',dx)
print('dt=',dt)
print('Number of iterations=',t_final/dt)

#we create arrays where we will store the time and the l2norm
l2norm_NCF=[]
l2norm_NCF.append(np.sqrt(np.sum(u_current_NCF**2)/len(u_current_NCF))) #We expect it to be a consant! 

time_NCF=[]
time_NCF.append(0.0)
```

```python
#create directory where to save images
print(os.getcwd())

os.makedirs('./images_' + str(run_name_NCF))

os.listdir('./')
```

```python
# Initilize time and iteration counter
t = 0.0
i = 0

#save the initial conditions
plt.plot(x, u_current_NCF)
plt.grid(c = 'gray', alpha = 0.3)
plt.title('Time='+str(round(t,2)))
plt.ylim(0,10.5)
plt.savefig('./images_' + str(run_name_NCF) + '/fig_'+str(i).zfill(5)+'.png', dpi=200)
plt.close()

#solve the advection equation
while t < t_final:
    u_next = u_current_NCF - (dt/dx)*(u_current_NCF)*(u_current_NCF - np.roll(u_current_NCF, 1))    
    
    # Update the solution
    u_current_NCF = u_next.copy()
    
    #Update the time step with the new velocity
    dt = cf*dx/(np.max(u_current_NCF))
    
    #advance the time 
    t += dt
    i += 1
    
    #compute the l2 norm and add the time to the time vector
    l2norm_NCF.append(np.sqrt(np.sum(u_current_NCF**2)/len(u_current_NCF)))
    time_NCF.append(t)
    
    #plot the current result and save in an image every 10 iterations
    if (i%5==0):
        plt.plot(x, u_current_NCF)
        plt.title('Time='+str(round(t,2)))
        plt.grid(c = 'gray', alpha = 0.3)
        plt.ylim(0,10.5)
        plt.savefig('./images_' + str(run_name_NCF) + '/fig_'+str(i).zfill(5)+'.png', dpi=200)
        plt.close()
```

```python
# Plot the final solution
plt.plot(x, u_initial, label='Initial')
plt.plot(x, u_current_NCF, label='Final')
plt.title(r'J=501')
plt.ylim(0,10.5)
plt.xlabel('x')
plt.ylabel('u')
plt.grid(color = 'gray', alpha = 0.3)
plt.legend()
#plt.show()
plt.savefig('i_f_NCF_'+str(run_name_NCF)'.png', dpi = 300)
```

```python
plt.plot(time_NCF,l2norm_NCF/l2norm_NCF[0])
plt.ylabel(r'$L_2$ norm')
plt.xlabel('Time')
plt.grid(c = 'gray', alpha = 0.3)
plt.show()
```
